## Palabras

```python
words = ['play', 'filling', 'bar', 'theatre', 'easygoing', 'date', 'lead', 'that', 'story',  'island']
```

**Imprime todas las letras en mayuscula**

```python
for i in range(len(words)):  
    words[i] = words[i].upper()

print(words)
```

## Numeros

**Crea una lista que contenga el cuadraro de cada numero entre 1 y 10**

```python
squared_numbers = []

for num in range(1, 11): 
    squared_numbers.append(num ** 2) 
print(squared_numbers)
```

**Imprime una lista que contenga el cuadrado de cada número impar del 1 al 10**

```python
squared_odds = []

for num in range(1, 11):  
    if num % 2 != 0:  
        squared_odds.append(num ** 2)  
print(squared_odds)
```

**Crear una lista con los cuadrados de todos los múltiplos de 8 por debajo de 1000**

```python
squared_eights = []
for num in range (8, 10001):
    if num % 8 == 0:
        squared_eights.append(num ** 2)
print(squared_eights)
```

## People

```python
people = [
    {
        "name": "Juan",
        "age": 34,
        "n_kids": 2
    },
    {
        "name": "Pepe",
        "age": 27,
        "n_kids": 0
    },
    {
        "name": "Sonia",
        "age": 41,
        "n_kids": 1
    },
    {
        "name": "Lucía",
        "age": 22,
        "n_kids": 2
    },
    {
        "name": "Leo",
        "age": 55,
        "n_kids": 5
    }
]
```

**Cuánta gente hay?**

```python
count = 0

for person in people:
    count += 1

print(count)
```

**Cuanta gente tiene hijos**?

```python
count = 0
for person in people:
    if person["n_kids"] > 0:
        count += 1
print(count)
```

**Cuantos hijos tienen en total?**

```python
count = 0
for person in people:
    count += person["n_kids"]
print(count)
```
